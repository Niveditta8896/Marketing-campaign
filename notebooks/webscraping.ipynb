```python
from bs4 import BeautifulSoup
import requests
import random
import pandas as pd
```

```python
url="http://books.toscrape.com"
```

```python
response=requests.get(url)
response
```

```python
soup=BeautifulSoup(response.text,"html")# response is the request and .text gets the raw data in the html format from the url
```

```python
books=soup.find_all("li",class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')# we specify class in html as class_
```

```python
titles,prices,availabilities,rating,lead_source=[],[],[],[],[]
lead_sources=['Facebook Ads', 'Google Ads', 'Affiliate', 'Newsletter Referral', 'Twitter Ads']
```

```python
for book in books:
    title=book.h3.a["title"]# from h3 tag under a tag extract tile attribute
    price=book.find("p",class_="price_color").text.strip().replace("Â","").replace("£","")
    # here we use.text to get the output from html tag  and .str is not a method and doest give us the text output 
    availability_tag = book.find("p", class_="instock availability")
    availability = availability_tag.text.strip() if availability_tag else "N/A"
    rating_tag=book.find("p")["class"][1]
    lead_source_tag=random.choice(lead_sources)# You're picking a random value from the list lead_sources
# And assigning it to each individual campaign during the loop
    
    titles.append(title)
    prices.append(price)
    availabilities.append(availability)
    rating.append(rating_tag)
    lead_source.append(lead_source_tag)
```

```python
df=pd.DataFrame({"campaign name":titles,"lead_sources":lead_source,"revenue":prices,"availabilities":availabilities,"ratings":rating})
```
